# Files, exceptional handling, logging and memory management Questions

1. Difference between interpreted and compiled languages  

Ans -

 Compiled Languages: The entire source code is translated into machine code by a compiler before execution (e.g., C, C++). This creates an independent executable file, leading to faster execution speeds but a slower development cycle.

 Interpreted Languages: The source code is translated and executed line-by-line at runtime by an interpreter (e.g., Python, JavaScript). This makes debugging easier and code highly portable, though it typically runs slower than compiled code.

 2. Exception handling in Python  

 Ans -  Exception handling is a programming mechanism used to respond to unexpected events (errors) that occur during execution, preventing the program from crashing abruptly. In Python, it allows a developer to intercept runtime errors, handle them gracefully, and keep the application running.

 3. Purpose of the finally block

 Ans -   The finally block is executed guaranteed, regardless of whether an exception was raised or caught inside the try-except structure. It is primarily used for clean-up actions, such as closing file streams, releasing network sockets, or closing database connections.

 4. Logging in Python

 Ans -   Logging is the process of tracking events that happen when software runs. Instead of using print() statements (which only write to standard output and are hard to manage in production), Python's built-in logging module provides a flexible framework to write diagnostic messages to files, consoles, or remote servers.

 5. Significance of the __del__ method

 Ans -   The __del__ method is known as a destructor. It is automatically invoked by Python when an object is about to be destroyed or garbage-collected. It is typically used to perform clean-up tasks that aren't handled by automatic memory management, such as closing open network connections.

 6. Difference between import and from... import

 Ans -

  a. import module_name: Imports the entire module namespace. You must use the module name as a prefix to access its contents (e.g., math.sqrt(4)).
   
  b. from module_name import elements: Imports specific variables, functions, or classes directly into the current namespace. You can use them directly without a prefix (e.g., sqrt(4)).

7. Handling multiple exceptions

Ans - We can handle multiple exceptions by adding multiple except blocks, or by grouping related exceptions together into a single tuple:

```python
try:

    val = 10 / int(input())
except ZeroDivisionError:
    print("Cannot divide by zero.")
except ValueError:
    print("Invalid input type.")
```

8. Purpose of the with statement  

Ans - The with statement leverages context managers to simplify resource management. When opening files, it ensures that the file stream is cleanly and automatically closed as soon as the block of code exits, even if exceptions or runtime errors occur inside the block.

9. Difference between multithreading and multiprocessing

Ans -

a. Multithreading: Runs multiple threads within a single process. They share the same memory space. Because of Python's Global Interpreter Lock (GIL), multithreading is optimal for I/O-bound tasks (like web scraping or file reading) but cannot run truly parallel on multiple CPU cores for heavy computing.

b. Multiprocessing: Spawns entirely separate processes, each running on its own CPU core with its own memory allocation. This bypasses the GIL entirely, making it ideal for CPU-bound computational tasks.

10. Advantages of using logging  

Ans - Persistent Records: Saves diagnostic tracking data directly to files for post-mortem analysis.Severity Levels: Easily categorizes issues into distinct severities (DEBUG, INFO, WARNING, ERROR, CRITICAL).Configurability: Allows turning off low-priority messages or changing output locations without rewriting application code.

11. Memory management in Python

Ans - Memory management involves how Python allocates and deallocates memory dynamically during execution. It utilizes a private heap space where all Python objects and data structures reside. The programmer does not need to allocate memory manually; the Python runtime handles it via internal management algorithms.

12. Basic steps involved in exception handling  

Ans -

a. try: Python monitors this block for any potential runtime errors.

b. except: Catches and processes specific exceptions if they occur inside the try block.

c. else: Executes optional code only if no exceptions were raised in the try block.

d. finally: Executes cleanup code unconditionally at the very end.

13. Importance of memory management

Ans - Efficient memory management prevents memory leaks (where memory is allocated but never freed), ensures that the application doesn't exhaust host system resources, and optimizes runtime performance by quickly recycling unused memory spaces.

14. Role of try and except

Ans -

a. try: Encapsulates code that has a risk of failing at runtime.

b. except: Defines the recovery or fallback code that triggers immediately if a designated error arises inside the companion try block.

15. How Python's garbage collection system works

Ans - Python manages memory automatically using two core strategies:

a.Reference Counting: Every object tracks how many variables point to it. When an object's reference counter drops to zero, its memory is instantly freed.

b. Cyclic Garbage Collector: Reference counting alone cannot handle reference cycles (e.g., Object A points to Object B, and Object B points to Object A). Python runs a generational garbage collector in the background to detect and destroy these isolated cycles.

16. Purpose of the else block in exception handling

Ans -   The else block runs only if the code inside the try block executed completely without throwing any exceptions. It is ideal for code that should only run upon a completely successful operation, keeping the try block focused strictly on risky statements.

17. Common logging levels

Ans - In increasing order of severity:

a. DEBUG: Detailed diagnostic details.

b. INFO: Confirmation that things are working as expected

c. WARNING: Indication that something unexpected happened (or a future problem may occur).

d. ERROR: A serious problem; the software couldn't perform a specific function.

e. CRITICAL: A fatal error; the program may be unable to continue running.

18. Difference between os.fork() and multiprocessing  

Ans -

a. os.fork(): A low-level system call specific to Unix/Linux operating systems. It clones the current process directly. It is not portable to Windows operating systems natively.

b. multiprocessing: A high-level, cross-platform built-in module. It abstracts process creation cleanly across Windows, Mac, and Linux, providing high-level tools like pools, queues, and pipes.

19. Importance of closing a file  

Ans -  Failing to close files can cause resource leaks, meaning the operating system keeps file descriptors locked. This can prevent other processes from accessing the file and might lead to data corruption or data loss if cached write buffers aren't fully flushed to the disk.

20. Difference between file.read() and file.readline()

Ans -

a. file.read(): Reads the entire content of the file from the current cursor position to the very end as a single string (or bytes object).

b. file.readline(): Reads only a single line from the file up to the next newline character (\n).

21. What the logging module is used for  

Ans - It is used to emit runtime log messages from Python applications. It allows developers to record diagnostic data to files, format messages uniformly with timestamps, and filter them systematically by importance levels.

22. What the os module is used for in file handling  

Ans - The os module provides operating-system-dependent functionality. In file handling, it is used to rename files, delete files, check file paths, construct cross-platform directories (os.path.join), and query metadata like file sizes.

23. Challenges associated with memory management

Ans -

a.  Reference Cycles: Objects referencing each other can stay stuck in memory if the cyclic collector isn't running frequently.

b. Memory Fragmentation: Memory may become fragmented over time, causing a high total memory footprint even if many internal objects have been freed.

c. Global Interpreter Lock (GIL) Overhead: Managing memory safely across multiple threads requires locking mechanisms that can introduce performance overhead.

6. Log an error message when division by zero occurs

```python
import logging

logging.basicConfig(filename="errors.log", level=logging.ERROR,
                    format="%(asctime)s - %(levelname)s - %(message)s")

try:
    result = 10 / 0
except ZeroDivisionError as e:
    logging.error("An error occurred: Division by zero attempted.", exc_info=True)
    print("An error occurred. Details have been saved to errors.log.")
```

15. Print content of a file and handle the case when it is empty

```python
import os

file_name = "manifest.txt"

try:
    if os.path.exists(file_name) and os.path.getsize(file_name) == 0:
        print("The file exists but is empty.")
    else:
        with open(file_name, "r") as file:
            content = file.read()
            if not content:
                print("The file contains no data.")
            else:
                print(content)
except FileNotFoundError:
    print("File not found.")
```

18. Implement basic logging setup with file rotation after 1MB

```python
import logging
from logging.handlers import RotatingFileHandler

handler = RotatingFileHandler("rotated.log", maxBytes=1000000, backupCount=3)

logging.basicConfig(level=logging.INFO,
                    format="%(asctime)s - %(levelname)s - %(message)s",
                    handlers=[handler])

logging.info("Rotating log environment configured.")
```

21. Read a file and print number of occurrences of a specific word

```python
target_word = "Python"
count = 0

try:
    with open("tutorial.txt", "r") as file:
        for line in file:

            words = line.split()
            for word in words:

                if word.strip(",.?!\"';:").lower() == target_word.lower():
                    count += 1
    print(f"The word '{target_word}' was detected {count} times.")
except FileNotFoundError:
    print("The requested file was not found.")
```
